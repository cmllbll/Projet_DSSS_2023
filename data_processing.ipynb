```python
# Import des packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv
from datetime import datetime
```

```python
### Lecture des données

# Initialisation de la liste de stockage
data = []

# Parcours du fichier csv pour les données 2015-2020
with open('data/fullcorpus20210725_DATE.csv', 'r', encoding='utf-8', errors='ignore') as file:
    reader = csv.reader(file)
    for line_number, line in enumerate(reader, start=1):
        try:
            data.append(line[1:])
        except Exception as e:
            print(f"Erreur à la ligne {line_number}: {e}")
df_2015_2020 = pd.DataFrame(data[1:],columns=data[0]).drop("crisis_term",axis=1)

# Lecture des données 2020-2022
df_2020_2022 = pd.read_csv(r'data/data_aug.csv',usecols=["Journal","Titre","Date","Texte","extrait_debut","extrait_fin","CJournal"])
df_2020_2022 = df_2020_2022.drop(df_2020_2022[df_2020_2022['Journal'] == 'de Dubaï'].index)
df_2020_2022["Auteur"] = None
df_2020_2022['Date'] = df_2020_2022['Date'].apply(lambda x: datetime.strptime(x, '%d/%m/%Y').strftime('%Y-%m-%d'))

# Final dataframe
df_final = pd.concat([
    df_2015_2020,
    df_2020_2022[df_2015_2020.columns]
])

# Ajustement des erreurs
df_final["CJournal"].replace("Autre", "Echos", inplace=True)
df_final["CJournal"].fillna("Parisien", inplace=True) # Un seul NA du journal Le Parisien

df_final.head()
```
